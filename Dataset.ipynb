```python
import json
import pandas as pd

data = []

with open("data/problems_data.jsonl", "r", encoding="utf-8") as f:
    for line in f:
        data.append(json.loads(line))

df = pd.DataFrame(data)
df.head()
```

```python
text_cols = [
    "title",
    "description",
    "input_description",
    "output_description"
]

for col in text_cols:
    if col in df.columns:
        df[col] = df[col].fillna("")
```

```python
df["full_text"] = (
    df.get("title", "") + " " +
    df.get("description", "") + " " +
    df.get("input_description", "") + " " +
    df.get("output_description", "")
)
```

```python
df["full_text"].iloc[0][:500]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(
    max_features=5000,
    stop_words="english"
)

X_text = tfidf.fit_transform(df["full_text"])

X_text.shape
```

```python
df["text_length"] = df["full_text"].apply(len)
```

```python
keywords = ["graph", "dp", "dynamic", "tree", "recursion"]

def keyword_count(text):
    text = text.lower()
    return sum(text.count(k) for k in keywords)

df["keyword_count"] = df["full_text"].apply(keyword_count)
```

```python
from scipy.sparse import hstack

X = hstack([
    X_text,
    df[["text_length", "keyword_count"]].values
])
```

```python
df.columns
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df["problem_class_encoded"] = le.fit_transform(df["problem_class"])
```

```python
df[["problem_class", "problem_class_encoded"]].head()
```

```python
y_class = df["problem_class_encoded"]
y_score = df["problem_score"]
```

```python
print(y_class.shape)
print(y_score.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, 
    y_class, 
    test_size=0.2,
    random_state=42
)
```

```python
from sklearn.linear_model import LogisticRegression

clf = LogisticRegression(
    max_iter=1000,
    n_jobs=-1
)

clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

acc = accuracy_score(y_test, y_pred)
print("Accuracy:", acc)
```

```python
from sklearn.svm import LinearSVC

clf = LinearSVC()
clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y_score,
    test_size=0.2,
    random_state=42
)
```

```python
from sklearn.ensemble import RandomForestRegressor

reg = RandomForestRegressor(
    n_estimators=200,
    random_state=42,
    n_jobs=-1
)

reg.fit(X_train, y_train)
```

```python
y_pred = reg.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))

print("MAE:", mae)
print("RMSE:", rmse)

```

```python
import joblib

joblib.dump(tfidf, "models/tfidf.pkl")
joblib.dump(clf, "models/classifier.pkl")
joblib.dump(reg, "models/regressor.pkl")
joblib.dump(le, "models/label_encoder.pkl")
```
